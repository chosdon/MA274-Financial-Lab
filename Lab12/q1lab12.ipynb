```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def P(t, T, r, a, b, sigma):
    tau = T - t
    B = (1 - np.exp(-a * tau)) / a
    A = np.exp((b - (0.5 * sigma**2) / a**2) * (B - tau) - (sigma**2 / (4 * a)) * B**2)
    return A * np.exp(-B * r)
```

```python
parameters = [ [5.9 ,0.2, 0.3, 0.1] , [3.9 ,0.1,0.3,0.2] , [0.1,0.4,0.11,0.1] ]
time = np.arange(1,11)

for para in parameters:
    a, b, sigma, r0 = para
    yeild = []
    for T in time:
        p = P(0, T, r0, a, b, sigma)
        y = -np.log(p) / T
        yeild.append(y)
    plt.plot(time, yeild, label=f"a={a}, b={b}, σ={sigma}, r0={r0}",color='green',marker='o')
    plt.title("Vasicek Yield Curves for parameters ")
    plt.xlabel("Maturity (T)")
    plt.ylabel("Yield")
    plt.legend()
    plt.grid(True)
    plt.show()
    
```

```python
Time = np.linspace(0.01, 10, 500, endpoint=False)
r_vals = [0.1 * i for i in range(1, 11)]

for i, para in enumerate(parameters):
    a,b,sigma , _ = para
    for r0 in r_vals:
        yield_vals = [] 
        for T in Time:
            p = P(0, T, r0, a, b, sigma)
            y = -np.log(p) / T
            yield_vals.append(y)
        plt.plot(Time, yield_vals )
        
    plt.title(f"Vasicek Yield Curves for 500 time steps for parameters {i+1} ")
    plt.xlabel("Maturity (T)")
    plt.ylabel("Yield")
    plt.legend()
    plt.grid(True)
    plt.show()
    

```
