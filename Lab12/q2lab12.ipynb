```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
parameters = [[0.02,0.7,0.02,0.1],
              [0.7,0.1,0.3,0.2],
              [0.06, 0.09, 0.5, 0.02]]
```

```python
def P(t, T, r0, a, b, sigma):
    tau = T - t
    gamma = np.sqrt(a*a + 2*sigma*sigma)

    num_B = 2 * (np.exp(gamma * tau) - 1)
    den_B = (a + gamma) * (np.exp(gamma * tau) - 1) + 2 * gamma
    B = num_B / den_B

    num_A = 2 * gamma * np.exp((a + gamma) * tau / 2)
    den_A = den_B  
    A = (num_A / den_A) ** (2 * a * b / (sigma * sigma))

    return A * np.exp(-B * r0)
```

```python

time = np.linspace(0.01, 10, 10, endpoint=False)

for i, para in enumerate(parameters):
    a, b, sigma, r0 = para 
    yields = []
    for T in time:
        p = P(0, T, r0, a, b, sigma)
        y = -np.log(p) / T  
        yields.append(y)
    
    plt.plot(time, yields, color='orange',marker = '*')
    plt.title(f'CIR model for parameters {i+1}')
    plt.xlabel('T (Maturity)')
    plt.ylabel('Yield')
    plt.grid(True)
    plt.show()
```

```python
Time = np.linspace(0.1, 600, num=600, endpoint=False)
r_vals = [0.1 * i for i in range(1, 11)]
a, b, sigma, _ = parameters[0]

for r in r_vals:
    yields = []
    for T in Time:
        p = P(0, T, r, a, b, sigma) 
        y = -np.log(p) / T      
        yields.append(y)
    plt.plot(Time, yields, )

plt.xlabel('Maturity (T)')
plt.ylabel('Yield')
plt.title('Term Structure for 10 values of r(0) and 600 time units')
plt.legend()
plt.grid(True)
plt.show()
```
